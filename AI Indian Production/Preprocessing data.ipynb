```python
import numpy as np 
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
train=pd.read_csv("House data/train.csv")
test=pd.read_csv("House data/test.csv")
print("The Shape Of Train Data :",train.shape)
print("The Shape Of Test Data :",test.shape)
```

```python
x_train=train.drop(columns=['SalePrice'],axis=1)
y_train=train['SalePrice']
x_test=test.copy()
```

```python
print("The Shape Of XTrain Data :",x_train.shape)
print("The Shape Of YTrain Data :",y_train.shape)
print("The Shape Of XTest Data :",x_test.shape)
```

```python
num_vars_mean=[num_vars_miss[0]]
num_vars_median=num_vars_miss[1:]
num_vars_mean,num_vars_median
```

```python
cat_vars_missing=cat_vars_miss[9:]
cat_vars_missing
```

```python
num_vars_mean_imputer=Pipeline(steps=[('imputer',SimpleImputer(strategy='mean'))])
num_vars_median_imputer=Pipeline(steps=[('imputer',SimpleImputer(strategy='median'))])
cat_vars_mode_imputer=Pipeline(steps=[('imputer',SimpleImputer(strategy='most_frequent'))])
cat_vars_missing_imputer=Pipeline(steps=[('imputer',SimpleImputer(strategy='constant',fill_value='chaws'))])
```

```python
preprocessor=ColumnTransformer(transformers=[('mean_imputer',num_vars_mean_imputer,num_vars_mean),
                                ("median_imputer",num_vars_median_imputer,num_vars_median),
                               ("mode_imputer",cat_vars_mode_imputer,cat_vars_mode),
                                ("missing_imputer",cat_vars_missing_imputer,cat_vars_missing)
                               ])
```
